```
# Import the required libraries
import random
import matplotlib.pyplot as plt

# Set the number of Iterations for the Simulations
n_iter=10000
```

```
def theoretical_ans(n):
  #Number of Events having 5 distinct faces
  n_event= 6*( (5)**n- 5*(4)**n+ 10*(3)**n- 10*(2)**n+ 5)
  # Corresponding Probability Values
  return n_event/(6**n)


# Function to run the simulation for n dice throws
def prob_vals_simuation(n):
  # event_count var keeps the count of the events where one of the faces does not show up
  event_count=0
  for i in range(n_iter):
    # sample_run stores the outcome of each of the n dice throws
    sample_run=[]  
    for j in range(n):
      sample_run.append(random.choice([1,2,3,4,5,6]))
    # If there are exactly 5 distinct faces in the n dice throws then increase the event_count by 1
    if len(set(sample_run))==5:
      event_count+=1
  # the corresponding probabilty value for the event
  return event_count/n_iter
```

```
# x_list denote the range of x values
x_list=range(1,21)


# Map function maps outcome of prob_vals_simuation function for each item in x_list
# Result from Simulation
y_sim=list(map(prob_vals_simuation,x_list))
# Result from Theoretical Answer
y_theor=list(map(theoretical_ans,x_list))
```

```
# Plot the two answers to see how different they are.
g1=plt.plot(x_list,y_sim,'bo')
g2=plt.plot(x_list,y_theor,'g-')
plt.xlabel('Number of Throws')
plt.ylabel('Probabilty Values')
plt.xticks(x_list)
legend=plt.legend((g1[0],g2[0]),('Simulation','Theoretical'))
plt.show()
```
