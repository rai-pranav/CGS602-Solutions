```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##Number of Samples to iterate over (M)
num_sample=[100,1000,10**5]

#Sample Sizes to iterate over (N)
sample_size=[2,10,20,40]

# Decription of the Population PDF
outcomes=np.array([1,2,9])
p_outcomes=[1.0/3.0,1.0/3.0,1.0/3.0]

# List to store the value of Sample Means
sample_means=[]

def draw_sampling_distr(seed):
    #random seed
    np.random.seed(seed)
    
    #Setting the number of subplots and figure dimensions etc.
    fig,arr = plt.subplots(len(sample_size),len(num_sample),figsize=(20, 10))
    fig.tight_layout(pad=3.0)

    
    ##Iterate over all the sample size and number of samples
    for i in range(0,len(sample_size)):
        for j in range(0,len(num_sample)):
            # Get random samples based on sample size and number of samples
            #See the documentation of np.random.choice here - https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html
            samples=np.random.choice(outcomes,(num_sample[j],sample_size[i]),p=p_outcomes)
            
            #Get the relevant statistic for the sampling distribution
            sample_stats=np.mean(samples,axis=1)
            #Set the title for the Sampling Distribution
            arr[i][j].set_title('N = '+ str(sample_size[i])+', M = '+str(num_sample[j]) +', \u03BC = {:.2f}'.format(np.mean(sample_stats)) +', \u03C3 = {:.2f}'.format(np.std(sample_stats)))
            arr[i][j].hist(sample_stats,bins=20,edgecolor='black', linewidth=1.2,density=True)
            
    plt.show()
```

```python
draw_sampling_distr(0)
```

# Conclusion -
1. As we increase the size of the samples , the distribution approaches Normal distribution. If the sample size is too low like 2, then increasing number of sample also does not lead to normal like distribution.
2. On increasing the sample size , the Standard Deviation of the Sampling Distribution (SEM) continues to decrease.
